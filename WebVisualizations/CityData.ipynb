```python
import pandas as pd
```

```python
df = pd.read_csv('./assets/cities.csv')

df['Date'] = pd.to_datetime(df['Date'], unit='s')

df['Date'] = df['Date'].dt.date
df.set_index('City_ID', inplace=True)

df.head()
```

```python
html = df.to_html()
```

```python
# Functions were provided by Shawn Stankevitsch
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False

def add_aligns(text):
    i=0
    j=len(text)
    k=[]
    reached_end = False
    while reached_end == False:
        a=text.find("<td>", i, j)
        b=text.find("</td>", i + 4, j)
        if is_number(text[a + 4:b]):
            k.append(a)
            i=b+5
        else:
            i=b+5
        if b == -1 or i>=j:
            reached_end = True
    for x in range(len(k)-1,-1,-1):
        text = text[:k[x] + 3] + ' align="right"' + text[k[x] + 3:]
    return text

print(add_aligns(html))
```
